## Cleanign/filtering the Data

```python
# -finding missing values
pkmn.isnull()
```

```python
# checking null value in each column
pkmn.isnull().sum()
```

```python
# droping the column which has high null value *cleaning*
newpkm = pkmn.drop(['Note'], axis=1)
newpkm.head()
```

```python
# again checking null value in each column for the new variable
newpkm.isnull()
```

```python
# checking new shape after cleaining
newpkm.isnull().shape
```

```python
# removing null value to NaN, for cleaning the data... rows and columns are reduced from 46,15
newpkm.dropna().shape
```

```python
# removing all missing value
newpkm = newpkm.dropna()
```

```python
newpkm.isnull().sum()
```
