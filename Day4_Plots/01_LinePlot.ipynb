## Important libraries for data visualisation
- seaborn automatically install these libraries

 **matplotlib*
 **numpy*
 **scipy*
 **pandas*

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool=sns.load_dataset("iris")
phool
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)  # don't write x= sepal length{no spacing b/w sepal & length}
plt.show()
```

## Adding title

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool=sns.load_dataset("iris")
phool
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)  # don't write x= sepal length{no spacing b/w sepal & length}
plt.title("peshawar is the city of flower")
plt.show()
```

## adding limits

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool=sns.load_dataset("iris")
phool
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)  # don't write x= sepal length{no spacing b/w sepal & length}
plt.title("peshawar is the city of flower")

# limit x and y axis
plt.xlim(4)
plt.ylim(1)

plt.show()
```

## set styles

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool=sns.load_dataset("iris")
phool
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)  # don't write like x= sepal length{no spacing b/w sepal & length}
plt.title("peshawar is the city of flower")

# makeing background black of chart with set style 
sns.set_style(style=None, rc=None)
sns.set_style("dark")

plt.show()
```

## Size of fiqure

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool=sns.load_dataset("iris")

# change fiqure size
plt.figure(figsize=(8,6))

sns.lineplot(x="sepal_length", y="sepal_width", data=phool)  # don't write x= sepal length{no spacing b/w sepal & length}
plt.title("peshawar is the city of flower")

plt.show()
```
